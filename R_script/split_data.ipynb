```python
import pandas as pd
```

```python
huffman_df = pd.read_csv('output_result_huffman_all_size_entropy.csv', index_col=0)
lzw_df = pd.read_csv('output_result_lzw_all_size_entropy.csv', index_col=0)
```

```python
# huffman_df = huffman_df.groupby('algo-format-size-repetation').mean().reset_index()
# lzw_df = lzw_df.groupby('algo-format-size-repetation').mean().reset_index()
```

```python
huffman_df['algo-format-size-repetation']
```

```python
huffman_df['algo'] = huffman_df['algo-format-size-repetation'].str.split('-').str[0]
huffman_df['format'] = huffman_df['algo-format-size-repetation'].str.split('-').str[1]
# huffman_df['size'] = huffman_df['algo-format-size-repetation'].str.split('-').str[2]
# huffman_df['repetation'] = huffman_df['algo-format-size-repetation'].str.split('-').str[3]

lzw_df['algo'] = lzw_df['algo-format-size-repetation'].str.split('-').str[0]
lzw_df['format'] = lzw_df['algo-format-size-repetation'].str.split('-').str[1]
# lzw_df['size'] = lzw_df['algo-format-size-repetation'].str.split('-').str[2]
# lzw_df['repetation'] = lzw_df['algo-format-size-repetation'].str.split('-').str[3]
```

```python
lzw_df
```

```python
og_df = pd.concat([huffman_df, lzw_df], axis=0)
og_df.columns
```

```python
og_df.loc[og_df["format"].isin(['txt', 'pdf']), "type"] = 'text'
og_df.loc[og_df["format"].isin(['jpg', 'png']), "type"] = 'image'
og_df.loc[og_df["format"].isin(['mp4', 'flv']), "type"] = 'video'
```

```python
og_df
```

```python
og_df['exp'] = og_df['algo-format-size-repetation']
og_df[['exp','algo', 'type', 'format', 'size', 'entropy', 'compression_ratio', 'compression_time','saving_size', 'total_power', 'cpu_utilization', 'energy_efficiency']].to_csv('output_result.csv', index=False)
```
